## Summary:
#### Air pollution poses a significant threat to public health and quality of life, particularly in densely populated and industrial regions of India. This project analyzes real-world air quality data collected from multiple cities and monitoring stations across India, provided by the Central Pollution Control Board (CPCB).

#### Using daily and hourly pollutant levels—including PM2.5, PM10, NO₂, SO₂, CO, and O₃—this study evaluates AQI trends, identifies pollution hotspots, and uncovers temporal patterns in air pollution (such as seasonal effects and time-of-day variations). The analysis also compares city-wise and station-wise AQI buckets (Good, Moderate, Poor, etc.), offering visual and statistical insights into the worsening air quality in urban centers like Delhi, Mumbai, and Patna.

#### PM2.5 – Particulate Matter ≤ 2.5 micrometers ,,,     PM10 – Particulate Matter ≤ 10 micrometers
#### NO – Nitric Oxide     NO2 – Nitrogen Dioxide  ,,,    NOx – Nitrogen Oxides (NO + NO2)
#### NH3 – Ammonia      CO – Carbon Monoxide    ,,,   SO2 – Sulfur Dioxide     O3 – Ozone    
#### Benzene – Benzene (a volatile organic compound)  ,,,  Toluene – Toluene (a volatile organic compound)
#### Xylene – Xylene (a volatile organic compound)  ,,,  AQI – Air Quality Index

## Datasets:

### seems like we got 50% of nulls in this city_day dataset, removing or dropping nulls will make this dataset useless. where we have to remove 50% of data. so just replace this mean values. mean values as per each city.

```python
df_City_day.dtypes
```

```python
numeric_cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']

for col in numeric_cols:
    df_City_day[col] = df_City_day.groupby('City')[col].transform(lambda x:x.fillna(x.mean()))
```

```python
 df_City_day.isnull().sum()
```

```python
### still we got some nulls. it may due to Nan values present in columns. 
### First we convert values to numeric
for col in numeric_cols:
    df_City_day[col] = pd.to_numeric(df_City_day[col], errors = 'coerce')

### Again filling with global mean values

for col in numeric_cols:
    df_City_day[col].fillna(df_City_day[col].mean(), inplace=True)
```

```python
 df_City_day.isnull().sum()
```

```python
df_City_day['AQI_Bucket'].value_counts()
```

```python
df_City_day['AQI_Bucket'].mode()
```

```python
## Lets Define a function to fill 0 value in the place of Nan values

def fill_mode(series):
    mode = series.mode()
    if not mode.empty:
        return series.fillna(mode[0])
    else:
        return series

### Apply it to each city group
df_City_day['AQI_Bucket'] = df_City_day.groupby('City')['AQI_Bucket'].transform(fill_mode)
```

```python
df_City_day.isnull().sum()
```

### now we got no null values in this df_city_day column, lets move to next.

```python
## Convert datetime column to datetime format
df_City_day['Date'] = pd.to_datetime(df_City_day['Date'], errors = 'coerce')

## Create new columns like day, month and year
df_City_day['Day'] = df_City_day['Date'].dt.day
df_City_day['Month'] = df_City_day['Date'].dt.month
df_City_day['Year'] = df_City_day['Date'].dt.year
```

```python
df_City_day[['Date', 'Day', 'Month', 'Year']].head()
```

```python
df_City_day.dtypes
```

```python
df_City_day.describe()
```

```python
df_City_day['AQI_Bucket'].value_counts()
```

```python
## AQI  by City
df_City_day.groupby('City')['AQI'].mean().sort_values(ascending = False)
```

### Ahmedabad, Delhi, Patna, Gurugram, Lucknow has highest value of Air Quality Index.
### Aizawl, Shillong, Coimbatore, Thiruvananthapuram and Ernakulam has lowest value.

```python
## Monthly AQI Trend
df_City_day.groupby(['Year','Month'])['AQI'].mean().unstack().plot(kind = 'line')
```

### Jan & November shows higher AQI levels.
### june to september shows lower AQI, due to monsoon helping pollutants.
### Year 2020 shows significant drop, maybe due to Covid.

```python
# Compute correlation matrix
corr_matrix = df_City_day[numeric_cols].corr()

# Show correlation of all numeric variables with AQI
aqi_corr = corr_matrix[['AQI']].drop('AQI')  # Drop AQI-AQI correlation (1.0)

plt.figure(figsize=(6, 8))
sns.heatmap(aqi_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation of AQI with Other Pollutants')
plt.show()
```

###  Feels like AQI influenced by PM2.5, PM10, NO, NO2, CO, SO2 strongly.

## Now lets move to station day and stations table.

```python
## Merge stations with station Day Table
df_station = pd.merge( df_station_day,
    df_stations,
    on='StationId',
    how='inner')

df_station.head(5)
```

```python
df_station.shape
```

```python
df_station.dtypes
```

```python
df_station.isnull().sum()
```

```python
df_station.describe()
```

### Again we got nulls. where column Xylene have about 85% of nulls of total value. we cant drop nulls. will affect our data.
### same like city_day column we fill nulls with mean values by stationid

```python
# List of numeric columns to fill
numeric_cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO',
                'SO2','O3','Benzene','Toluene','Xylene','AQI']

# Fill nulls with mean per station
for col in numeric_cols:
    df_station[col] = df_station.groupby('StationId')[col].transform(lambda x: x.fillna(x.mean()))
```

```python
df_station.isnull().sum()
```

### see again facing nulls, which may due to nan values, will fill by Global Mean Filling (To handle remaining NaNs)

```python
df_station[numeric_cols] = df_station[numeric_cols].fillna(df_station[numeric_cols].mean())
```

```python
df_station.isnull().sum()
```

```python
df_station['AQI_Bucket'].value_counts()
```

```python
df_station['Status'].value_counts()
```

```python
df_station['AQI_Bucket'] = df_station.groupby('StationId')['AQI_Bucket'].transform(
    lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x)
```

```python
df_station['Status'] = df_station['Status'].fillna('Active')
```

```python
df_station.isnull().sum()
```

```python
# Fill remaining nulls with overall mode
df_station['AQI_Bucket'].fillna(df_station['AQI_Bucket'].mode()[0], inplace=True)
```

```python
df_station.isnull().sum()
```

```python
df_station['Date'] = pd.to_datetime(df_station['Date'])
df_station['Day'] = df_station['Date'].dt.day
df_station['Month'] = df_station['Date'].dt.month
df_station['Year'] = df_station['Date'].dt.year
```

```python
df_station.head(5)
```

```python
## Top / Bottom polluted Stations
top_stations = df_station.groupby('StationName')['AQI'].mean().sort_values(ascending=False).head(10)
top_stations.plot(kind='bar', title='Top 10 Most Polluted Stations (Avg AQI)')
```

```python
top_stations = df_station.groupby('StationName')['AQI'].mean().sort_values(ascending=True).head(10)
top_stations.plot(kind='bar', title='Bottom 10 Most Polluted Stations (Avg AQI)')
```

### Delhi, Ahmedabad and Lucknow are in top and Mizoram, kerala, coimbatore etc are in bottom.

```python
# Monthly AQI trend (across all stations)
monthly_trend = df_station.groupby(['Year', 'Month'])['AQI'].mean().reset_index()
sns.lineplot(data=monthly_trend, x='Month', y='AQI', hue='Year')
plt.title('Monthly AQI Trend')
```

### see 2020 has very low AQI due to covid, 2015 to 2017 has high AQI value and there is high spike in AQI value, during winter season due to stubble burning, low wind and DIWALI.

#### see we got many nulls, can't fill or drop entire nulls. so just take only 3 columns (City, Datetime, AQI)

```python
df_cHour_AQI = df_City_hour[['City','Datetime','AQI']]
df_cHour_AQI.isnull().sum()
```

```python
df_cHour_AQI.shape
```

#### AQI nulls are very low compared to total rows, so just remove nulls based on AQI column.

```python
df_cHour_AQI = df_cHour_AQI.dropna(subset = "AQI")
df_cHour_AQI.isnull().sum()
```

```python
df_cHour_AQI.head(5)
```

```python
df_cHour_AQI['Datetime'] = pd.to_datetime(df_cHour_AQI['Datetime'], errors = 'coerce')

df_cHour_AQI['Day'] = df_cHour_AQI['Datetime'].dt.day
df_cHour_AQI['Month'] = df_cHour_AQI['Datetime'].dt.month
df_cHour_AQI['Year'] = df_cHour_AQI['Datetime'].dt.year
df_cHour_AQI['Hour'] = df_cHour_AQI['Datetime'].dt.hour

df_cHour_AQI.head(5)
```

```python
df_cHour_AQI.dtypes
```

```python
df_cHour_AQI['Month_name'] = df_cHour_AQI['Datetime'].dt.strftime('%b')
df_cHour_AQI.head(5)
```

```python
Ahmedabad_df = df_cHour_AQI[df_cHour_AQI['City'] == 'Ahmedabad'].copy()
Hourly_AQI = Ahmedabad_df.groupby('Hour')['AQI'].mean().reset_index()

plt.figure(figsize = (12,6))
sns.lineplot(data = Hourly_AQI, x = 'Hour', y = 'AQI', marker='o', color='blue')
plt.title('Average Hourly AQI in Ahmedabad', fontsize=16)
plt.xlabel('Hour of Day (0–23)', fontsize=12)
plt.ylabel('Average AQI', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

### AQI remains high from midnight to 4am at 650, for 9am to 17pm seems something low like ~250, after 7pm it spikes to 600.
### It may due to continous emissions at night, low winds, night time traffics, less atmospheric dispersion
